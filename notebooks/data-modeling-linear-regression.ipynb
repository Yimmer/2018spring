# 数据建模举例：线性回归

## 产生训练数据

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def get_linear(x, k=1, b=10, noise=1):
    y = k*x + b + noise * np.random.normal(0, 0.5, len(x))
    return y

X_train = np.linspace(0, 10, 101)
Y_train = get_linear(x)
Y = get_linear(X_train, noise=0)

fig, ax = plt.subplots()
lines = ax.plot(X_train, Y, 'k-', X_train, Y_train, 'r.')
ax.legend(lines, ['target', 'observed'])
```

## 线性回归

```python
from sklearn import linear_model
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

X_train = X_train.reshape(-1,1) # Expected 2D array
regr.fit(X_train, Y_train)
print('Coefficients: %g, intercept %g' % (regr.coef_, regr.intercept_ ))
#print regr.get_params()
Y_estimated = get_linear(X_train, k=regr.coef_, b=regr.intercept_, noise=0)

fig, ax = plt.subplots()
lines = ax.plot(X_train, Y, 'k-', X_train, Y_estimated, 'r--')
ax.legend(lines, ['true', 'estimated'])
```
